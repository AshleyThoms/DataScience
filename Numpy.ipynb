```python
# Python List vs Numpy

# how much execution time will be taken by List and sum()

import time
numbers = __builtins__.list(range(1,100000001))
total = 0
start_time = time.time()    # record start time
for i in numbers:
  total = total + i
end_time = time.time()     # record end time
print('Total :', total)
print("Execution Time : ",end_time - start_time,"seconds")
print(f"Execution Time : {end_time - start_time:.4f} seconds")
#.4f means show 4 digits after the decimal point
```

```python
numbers = np.arange(1,100000001)
total = 0
start_time = time.time()  # record start time
total = np.sum(numbers)
end_time = time.time()  # record end time
print("Total : ",total)
print("Execution Time : ",end_time - start_time,"seconds")
print(f"Execution Time : {end_time - start_time:.4f} seconds")
```

```python
# copy() vs view() method of numpy

# copy()
# a copy will create a new independent array in a different memory location

org_array = np.array([1,2,3,4,5,6])
print(org_array)

copied_array = org_array.copy()
print(copied_array)

# if we make any changes in copied array, it wont effect the original array
copied_array[0] = 100
print(copied_array)
print(org_array)
```

```python
# View()
# a view is the representation of the original array

org_array = np.array([1,2,3,4,5,6])
print(org_array)

copied_arr = org_array.view()
print(copied_arr)

# if we make any changes in copied array, it will effect the original array
copied_arr[0] = 100
print(copied_arr)
print(org_array)

# if we make any changes in org array, it will effect the copied array
org_array[4] = 400
print(org_array)
print(copied_arr)
```

```python
org_array = np.array([1,2,3,4,5,6])
copied_arr = org_array.view()

# shares_memory() --> return True if arrays are sharing the same memory location

print(np.shares_memory(org_array,copied_arr))
```

```python
prices = np.array([100,20,3,30,90,100,11,88,98,45])
# show the even number prices
newarr = prices[prices%2==0]
print(newarr)
```

```python
prices = np.array([100,20,3,30,90,100,11,88,98,45])
# np.where(expression)
# np.where returns INDEX NUMBERS
res = np.where(prices%2==0)
print(prices[res])
```

```python
prices = np.array([100,20,3,30,90,100,11,88,98,45])
res = np.where((prices%2==0) & (prices>=50))
print(prices[res])
# & combines both condition
```

```python
prices = np.array([100,20,3,30,90,100,11,88,98,45])
print(prices[(prices%2==0) & (prices>=50)])
```

```python
# SALARY CLASSIFICATION
salary = np.array([25000, 40000, 55000, 70000, 30000, 90000])
# Replace salary with High if salary >= 60000
# Replace salary with Medium if salary >= 40000 and <= 60000
# Otherwise Low

# np.where(condition,value_if_true,value_if_false)

result = np.where(
    salary >= 60000 , "High" ,
    np.where(salary >=40000 , "Medium" ,"Low")
)

print(result)
```
