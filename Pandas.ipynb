```python
# Pandas Data Science Library
import numpy as np
import pandas as pd
print(pd.__version__)
```

```python
# numpy --> array()
arr = np.array([10,20,30,40,50])
print(type(arr))
print(arr)
```

```python
# Pandas Series
# Series is 1D Labelled Array

s = pd.Series([10,20,30,40,50])
print(type(s))
print(s)
print(s.shape)

# In numpy --> Only Numbers
# In series --> Numbers + Labels (index)
```

```python
info = pd.Series(['John', 26, 6700, 'Klang'], index = ['Name', 'Age', 'Salary', 'City'])
print(info)
print(info['Name']) # Indexing
print(info.get('City')) # Getting data values using get()
```

```python
info = pd.Series(['John', 26, 6700, 'Klang'], index = ['Name', 'Age', 'Salary', 'City'])
print(type(info))
print(info['Name'])  # indexing
print(info.get('City')) # Getting data values using get()
print(info[0])  # iloc --> index location

# Series is like a dictionary + array combo
```

```python
# Indexing Slicing And Using For Loop
s = pd.Series([100,200,300,400,500, 600], index=['v1', 'v2', 'v3', 'v4', 'v5', 'v6'])
print(s)

# 100 200 300 400 500 600
# -6  -5  -4  -3  -2  -1
#v1  v2   v3  v4  v5  v6

# Indexing in Series can be done by using [] or iloc[] or Loc[]
print(' \nIndexing In Series')
print(s.iloc[0])
# for index numbers
print(s.loc['v5'])
# Loc used for Labels
print(s['v2'])
# [] used for Labels
print (s.iloc[-1])

print('\nSlicing In Series')
print(s[1:4]) # 4 will be excluded --> stopIndex
print(s['v1':'v5']) # if performing slicing using labels, the end label will be included

# 100 400 600
print(s[['v1','v4','v6']]) # for specific information s[['','','']]
print(s[['v1','v4','v6']].values)

# conditional selection
print('\nConditional In Series')
print(s[s%3==0])
```

```python
# using for loop
for i in s:
  print(i)
```

```python
for k,v in s.items():
  print(k,v)
```

```python
for i in enumerate(s):
  print(i)
```
